```python
# Tic Tac Toe

import random

def drawBoard(board):
    # Ham nay don gian la in ra 1 bang voi kich thuoc 3x3
    # Bat dau tu board[1]-> board[9].
    # board truyen vao la 1 list
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-'*12)
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-'*12)
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')

def inputPlayerLetter():
    # Nguoi choi lua chon 'X' hoac 'O'
    # Ham nay tra lai 1 list chua 'X' va 'O', Neu nguoi choi lua chon 'X' thi may tinh se
    # co lua chon la 'O'
    letter = ''
    while not (letter == 'X' or letter == 'O'):
        print("Do you want to be X or O?")
        letter = input().upper() #tat ca se duoc in Hoa
    if letter == 'X':
        return ['X', 'O']
    else:
        return ['O', 'X']

def whoGoesFirst():
    # Lua chon 1 cach ngau nhien xem nguoi choi hay may tinh di truoc
    if random.randint(0,1) == 0:
        return 'computer'
    else:
        return 'player'

def playAgain():
    # Ham nay tra lai True neu nguoi choi muon choi lai, nguoc lai se tra lai False
    print("Do you want to play again?(y or n)")
    return input().lower().startswith('y') #Kiem tra xem nguoi dung nhap vao co bat dau bang chu 'y' hay khonh

def makeMove(board, letter, move):
    board[move] = letter

def isWinner(bo, le):
    #  Kiem tra dieu kien thang
    # Tra lai True neu thang va nguoc lai
    return ((bo[7] == le and bo[8] == le and bo[9] == le) or
            (bo[4] == le and bo[5] == le and bo[6] == le) or
            (bo[1] == le and bo[2] == le and bo[3] == le) or
            (bo[7] == le and bo[4] == le and bo[1] == le) or
            (bo[8] == le and bo[5] == le and bo[2] == le) or
            (bo[9] == le and bo[6] == le and bo[3] == le) or
            (bo[7] == le and bo[5] == le and bo[3] == le) or
            (bo[9] == le and bo[5] == le and bo[1] == le))
def getBoardCopy(board):
    # Tao ra 1 nhan ban cua board va tra ve nhan ban do
    dupeBoard = []
    for i in board:
        dupeBoard.append(i)
    return dupeBoard

def isSpaceFree(board, move):
    # Tra lai tru neu vi tri board[move] con trong o tren board
    return board[move] == ' '

def getPlayerMove(board):
    # Nguoi choi chon vi tri tren bang de dat X or O
    move = ' '
    while move not in '1 2 3 4 5 6 7 8 9'.split() or not isSpaceFree(board, int(move)):
        print("What is your next move?(1-9)")
        move = input()
    return int(move)

def chooseRandomMoveFromList(board, movesList):
    # Tra ve 1 vi tri co the di chuyen o tren bang
    # Tra ve None neu khong con vi tri nao co the di chuyen
    possibleMoves = []
    for i in movesList:
        if isSpaceFree(board, i):
            possibleMoves.append(i)

    if len(possibleMoves) != 0:
        return random.choice(possibleMoves)
    else:
        return None

def getComputerMove(board, computerLetter):
    # Dua ra 1 bang va lua chon cua may tinh('X' or 'O'), xac dinh noi di chuyen va tra ve vi tri do
    if computerLetter == 'X':
        playerLetter = 'O'
    else:
        playerLetter = 'X'

    # Day la thuat toan cho Tic Tac Toe AI
    # Kiem tra xem nuoc di tiep theo co the thang hay khong
    for i in range(1,10):
        copy = getBoardCopy(board)
        if isSpaceFree(copy, i):
            makeMove(copy, computerLetter, i)
            if isWinner(copy, computerLetter):
                return i
    # Kiem tra xem nguoi choi co the thang o nuoc di tiep theo hay khong va chan nuoc di do
    for i in range(1,10):
        copy = getBoardCopy(board)
        if isWinner(copy, i):
            makeMove(copy, playerLetter, i)
            if isWinner(copy, playerLetter):
                return i
    # Chon 1 trong cac goc neu con trong
    move = chooseRandomMoveFromList(board, [1, 3, 7, 9])
    if move != None:
        return move

    #Chon vi tri trung tam neu con trong
    if isSpaceFree(board, 5):
        return 5

    # Chon 1 trong cac mat ben
    return chooseRandomMoveFromList(board, [2, 4 ,6, 8])

def isBoardFull(board):
    # Tra lai True neu toan bo bang khong con trong, va nguoc lai
    for i in range(1,10):
        if isSpaceFree(board, i):
            return False
    return True

# ----------- Bat dau vao doan de chay chuong trinh ---------------
print("-------------------Welcome to Tic Tac Toe--------------")
while True: # Vong lap vo tan
    # Tao bang rong
    theBoard = [' '] * 10
    playerLetter, computerLetter = inputPlayerLetter()
    turn = whoGoesFirst()
    print('The '+turn+' will go first.')
    gameIsPlaying = True # tao 1 flag

    while gameIsPlaying:
        if turn == 'player':
            # Luot choi cua nguoi choi
            drawBoard(theBoard)
            move = getPlayerMove(theBoard)
            makeMove(theBoard, playerLetter, move)

            if isWinner(theBoard, playerLetter):
                drawBoard(theBoard)
                print('Chuc mung ! Ban da thang !')
                gameIsPlaying = False

            else:
                if isBoardFull(theBoard):
                    drawBoard(theBoard)
                    print('Hoa roi !')
                    break

                else:
                    turn = 'computer'

        else:
            #Luot choi cua may
            move = getComputerMove(theBoard, computerLetter)
            makeMove(theBoard, computerLetter, move)

            if isWinner(theBoard, computerLetter):
                drawBoard(theBoard)
                print('May tinh da danh bai ban !')
                gameIsPlaying = False

            else:
                if isBoardFull(theBoard):
                    drawBoard(theBoard)
                    print('Hoa roi!')
                    break

                else:
                    turn = 'player'
    if not playAgain():
        break
```
